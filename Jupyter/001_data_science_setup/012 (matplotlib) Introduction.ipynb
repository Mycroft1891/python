# Introduction to Matplotlib

In order to display matoplotlib's generated plots in our notebook we need to use a magic command called `%matplotlib inline`.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot();
```

```python
x = [2,4,6,8]
y = [10,20,30,40]

plt.plot(x,y);
```

There are 3 methods of ploting data

```python
figure = plt.figure()
ax = figure.add_subplot()
plt.show()
```

```python
figure = plt.figure()
ax = figure.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
# prefered way
figure, ax = plt.subplots()
ax.plot(x,[30, 45, 80, 90]);
```
