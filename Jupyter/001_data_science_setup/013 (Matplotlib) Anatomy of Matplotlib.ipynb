# Anatomy of a Figure

When working with matplotlib a common workflow for creating a figure would be:

1. import matplotlib
2. prepare our data
3. setup the plot
4. customize the plot
5. save and show the plot

```python
import matplotlib.pyplot as plt
```

```python
x = [2,6,10,15]
y = [33,76,80,110]
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x,y);
```

```python
ax.set(
    title="Our plot",
    xlabel="My fancy X axis",
    ylabel="Your fancy Y axis"
)
```

```python
fig.savefig("_images/our_fancy_plot.png")
```
