# Anatomy of a Figure

When working with matplotlib a common workflow for creating a figure would be:

1. import matplotlib
2. prepare our data
3. setup the plot
4. customize the plot
5. save and show the plot

```python
import matplotlib.pyplot as plt
```

```python
x = [2,6,10,15]
y = [33,76,80,110]
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x,y);
```

```python
ax.set(
    title="Our plot",
    xlabel="My fancy X axis",
    ylabel="Your fancy Y axis"
)
```

```python
fig.savefig("_images/our_fancy_plot.png")
```

## Figures with Numpy arrays

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
x[:10]
```

```python
fig, ax = plt.subplots()
ax.plot(x,x**2);
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.cos(x));
```

```python
book_sales = {
    "1984": 12400,
    "Do Androids dream of Electric Sheep": 5400,
    "Neuromancer": 6000
}
```

```python
fig, ax = plt.subplots()
ax.bar(
    book_sales.keys(),
    book_sales.values()
);
ax.set(
    title="Book Sales by Novel",
    xlabel="Novel Title",
    ylabel="# of Sales",
);
```

```python
fig, ax = plt.subplots();
ax.barh(
    list(book_sales.keys()),
    list(book_sales.values())
)
ax.set(
    title="Book Sales by Novel",
    xlabel="Novel Title",
    ylabel="# of Sales",
);
```

```python
normal_distribution = np.random.randn(1000)

fig, ax = plt.subplots()
ax.hist(normal_distribution);
```

## Making multiple axis

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(15,15)
)

ax1.plot(x,np.exp(x))
ax2.bar(x,np.tan(x))
ax3.scatter(x,np.sin(x))
ax4.hist(x)
```
